```python
import torch
import torchvision
import torchvision.transforms as tr
from torch.utils.data import DataLoader, Dataset
import numpy as np
```

```python
transf = tr.Compose([tr.Resize(28),tr.ToTensor()])
```

```python
trainset = torchvision.datasets.MNIST('./data',train=True, download = True, transform=transf)
testset = torchvision.datasets.MNIST('./data',train=True, download = True, transform=transf)
```

```python
trainset[0] #3차원으로 이루어졌으며, 첫번째는 데이터값 두 번째는 라벨 값이다. [[[data,],[data,]]],label이다. 
#1*28*28
```

```python
trainset[0][0].size()
#Gray_Image이미지이므로 1개의 채널값으로 이루어짐
```

```python
#MNIST의 이미지는 총 60,000장임 배치사이즈는 50으로 설정함. 총 1200개의 배치가 생성 됨
trainloader = DataLoader(trainset, batch_size=50, shuffle=True)
testloader = DataLoader(testset, batch_size=50, shuffle=False)
len(trainloader)
```

```python
dataiter = iter(trainloader) #데이터를 하나 확인하기 위해 dataiter변수에다가 반복가능한 변수로 초기화 시키는 iter 함수
images, labels = dataiter.next() #반복가능해진 dataiter 변수의 첫 데이터를 추출하는 next함수 

print(images.size()) 
#배치크기 X 채널 X 너비 X 높이
#총 50개의 이미지(채널 1개, 너비 28, 높이 28)를 확인함.
```
